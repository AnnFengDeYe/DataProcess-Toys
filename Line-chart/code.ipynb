```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import numpy as np

# 
font_path = '/Users/annfengdeye/Library/Fonts/SimHei.ttf'  # 
font = FontProperties(fname=font_path, size=12)

# 
years = [2019, 2020, 2021, 2022, 2023]
debt_ratio = [0.267, 0.085, 0.077, 0.06, 0.078]
current_ratio = [1.88, 10.22, 9.74, 14.43, 11.05]
quick_ratio = [1.88, 10.22, 9.74, 14.42, 11.05]
cash_ratio = [0.101, 0.0178, 0.027, 0.042, 0.03]

# 
fig, ax = plt.subplots(figsize=(12, 7))  # 

# 
ax.plot(years, debt_ratio, label='', marker='o', color='#4c72b0')
ax.plot(years, current_ratio, label='', marker='s', color='#dd8452')
ax.plot(years, quick_ratio, label='', marker='^', color='#55a868')
ax.plot(years, cash_ratio, label='', marker='*', color='#c44e52')


# 
ax.set_title('', fontsize=16, fontproperties=font)
ax.set_xlabel('', fontsize=12, fontproperties=font)
ax.set_ylabel(' (%)', fontsize=12, fontproperties=font)  # "%"

# 
ax.set_xticks(years)
# y
# max_y = max(max(current_ratio), max(quick_ratio))
# ax.set_yticks(np.arange(0, max_y + 1, 2))  # 

# 
handles, labels = ax.get_legend_handles_labels()
for i, label in enumerate(labels):
    labels[i] = ax.get_legend_handles_labels()[1][i]

ax.legend(handles, labels, loc='upper right', bbox_to_anchor=(1, 0.7), fontsize=10, prop=font)


# 
ax.grid(True, linestyle='--', alpha=0.5)

# 
plt.tight_layout()

# PNG
plt.savefig('solvency_chart.png', dpi=300)

# 
plt.show()
```
