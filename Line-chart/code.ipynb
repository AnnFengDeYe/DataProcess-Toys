```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

# 假设mac中的字体地址为xxx.ttf  (请替换为实际路径)
font_path = '/Users/annfengdeye/Library/Fonts/SimHei.ttf'

# 创建FontProperties对象
font = FontProperties(fname=font_path, size=12)  # 可以根据需要调整字体大小

# 数据
years = [2019, 2020, 2021, 2022, 2023]
gross_margin = [37.30, 37.80, 36.60, 34.50, 37.30]
roa = [16.51, 15.01, 15.69, 8.88, 7.53]
roe = [24.42, 17.20, 16.79, 10.64, 9.97]

# 创建图形和坐标轴
fig, ax = plt.subplots(figsize=(10, 6))  # 调整图形大小

# 绘制折线图
ax.plot(years, gross_margin, label='毛利率', marker='o', color='#4c72b0')
ax.plot(years, roa, label='总资产净利率', marker='s', color='#dd8452')
ax.plot(years, roe, label='净资产收益率', marker='^', color='#55a868')

# 添加标题和标签 (使用fontproperties参数指定字体)
ax.set_title('', fontsize=16, fontproperties=font)
ax.set_xlabel('年份', fontsize=12, fontproperties=font)
ax.set_ylabel('数值 (%)', fontsize=12, fontproperties=font)


# 设置坐标轴刻度
ax.set_xticks(years)
ax.set_yticks(range(0, 41, 5))

handles, labels = ax.get_legend_handles_labels()  # 获取图例的 handles 和 labels
for i, label in enumerate(labels):
    labels[i] = ax.get_legend_handles_labels()[1][i]
# ax.legend(handles, labels, loc='upper right', fontsize=10, prop=font)  # prop参数设置图例字体
# 使用bbox_to_anchor和loc微调位置
ax.legend(handles, labels, loc='upper right', bbox_to_anchor=(1, 0.8), fontsize=10, prop=font)


# 添加网格线
ax.grid(True, linestyle='--', alpha=0.5)

# 美化图形
plt.tight_layout()

# 保存为PNG图片
plt.savefig('profitability_chart.png', dpi=300)

# 显示图形 (可选)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import numpy as np

# 字体设置
font_path = '/Users/annfengdeye/Library/Fonts/SimHei.ttf'  # 替换为您的字体路径
font = FontProperties(fname=font_path, size=12)

# 数据
years = [2019, 2020, 2021, 2022, 2023]
debt_ratio = [0.267, 0.085, 0.077, 0.06, 0.078]
current_ratio = [1.88, 10.22, 9.74, 14.43, 11.05]
quick_ratio = [1.88, 10.22, 9.74, 14.42, 11.05]
cash_ratio = [0.101, 0.0178, 0.027, 0.042, 0.03]

# 创建图形和坐标轴
fig, ax = plt.subplots(figsize=(12, 7))  # 调整图形大小

# 绘制折线图
ax.plot(years, debt_ratio, label='资产负债率', marker='o', color='#4c72b0')
ax.plot(years, current_ratio, label='流动比率', marker='s', color='#dd8452')
ax.plot(years, quick_ratio, label='速动比率', marker='^', color='#55a868')
ax.plot(years, cash_ratio, label='现金比率', marker='*', color='#c44e52')


# 添加标题和标签
ax.set_title('', fontsize=16, fontproperties=font)
ax.set_xlabel('年份', fontsize=12, fontproperties=font)
ax.set_ylabel('数值 (%)', fontsize=12, fontproperties=font)  # 去掉"%"，因为有些指标不是百分比

# 设置坐标轴刻度
ax.set_xticks(years)
# 设置y轴范围和刻度（根据数据范围调整）
# max_y = max(max(current_ratio), max(quick_ratio))
# ax.set_yticks(np.arange(0, max_y + 1, 2))  # 根据最大值设置刻度

# 添加图例
handles, labels = ax.get_legend_handles_labels()
for i, label in enumerate(labels):
    labels[i] = ax.get_legend_handles_labels()[1][i]

ax.legend(handles, labels, loc='upper right', bbox_to_anchor=(1, 0.7), fontsize=10, prop=font)


# 添加网格线
ax.grid(True, linestyle='--', alpha=0.5)

# 美化图形
plt.tight_layout()

# 保存为PNG图片
plt.savefig('solvency_chart.png', dpi=300)

# 显示图形
plt.show()
```
